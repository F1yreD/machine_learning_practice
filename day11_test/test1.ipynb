```python
import pandas as pd
import numpy as np
```

### 1、练习python列表，字典转series与df

```python
dic = {'a': 1, 'b': 2, 'c': 3}
s = pd.Series(dic)
print(s)
df = pd.DataFrame([dic])
print(df)
```

### 2、练习series和df的索引操作

```python
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
print(s, s.loc['b'], s.iloc[1], s.loc[['a', 'b']], s.iloc[[0, 1]], sep='\n'*2)
```

```python
df = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]},
                  index=['001', '002', '003'])
print(df, df.loc['001'], df.loc[['001', '002']], df.iloc[0],
      df.iloc[0:2], sep='\n'*2, end='\n'*2)  # row
print(df['a'], df[['a', 'b']], sep='\n'*2, end='\n'*2)  # col
print(df.iloc[1, 1], df.iloc[[1, 2], [0, 1]])  # loc
```

### 3、完成重复值去除，缺失值删除，缺失值填充练习

```python
data = {
    'a': ['a1', 'a2', 'a1'],
    'b': ['b1', 'b2', 'b3']
}
df = pd.DataFrame(data)
print(df, end='\n\n')

print(df.duplicated(subset=['a']))
df_col = df.drop_duplicates(subset=['a'], keep='first')
print(df_col)
```

```python
data = {
    'a': ['a1', np.nan, 'a3'],
    'b': [1, 2, 3]
}
df = pd.DataFrame(data)
print(df, end='\n\n')

print(df.isnull(), df.isnull().sum(axis=1), sep='\n',end='\n\n')
df_drop = df.dropna(axis=0)
print(df_drop)
df_fill_na = df.fillna('na')
print(df_fill_na)
```

### 4、完成apply和map的练习

```python
df = pd.DataFrame(np.arange(6).reshape(2, 3))
print(df)
df_row_mean = df.apply(lambda x: x.mean(),axis=0)
print(df_row_mean)
print('-'*20)
a = 0
df_loc_plus = df.map(lambda x:x+df_row_mean[a])
print(df_loc_plus)
```

### 5、完成groupby与agg的练习

```python
data = {
    "name": ["Alice", "Alice", "Charlie", "David", "Alice", "Charlie"],
    "age": [20, 22, 21, 20, 20, 21],
    "score": [85, 90, 78, 92, 85, 78],
    "gender": ["女", "男", "男", "男", "女", "男"]
}
df = pd.DataFrame(data, index=['s1', 's2', 's3', 's4', 's5', 's6'], columns=['name', 'age', 'score', 'gender'])
print(df)
print(df.groupby('name').agg({'age': ['mean', 'max'], 'score': 'sum'}))
```


### 6、编写机器学习的训练过程，什么是特征，什么是标签，什么是训练集，什么是测试集，什么是有监督，无监督学习，什么是分类问题，回归问题，什么是欠拟合和过拟合，以问答题写到代码文件里即可。
1. 特征：单条记录的属性值
2. 标签：样本的预测目标值
3. 训练集：用来训练模型的样本，包含特征和标签
4. 测试集：用来评估模型性能的样本，标签不放入模型运行
5. 有监督学习：使用有标签的样本训练，输出标签
6. 无监督学习：不输入标签训练，进行聚类、降维的任务
7. 分类问题：将样本划分进已有的类别
8. 回归问题：预测一个具体的数据
9. 欠拟合：随着模型复杂度增加，训练误差还在逐渐变小，测试误差逐渐变小
10. 过拟合：随着模型复杂度增加，训练误差不断降低，但测试误差反而增大
