### 7、Numpy youtube数据读取练习，并通过matplotlib绘制直方图

```python
us_path = "./youtube_video_data/US_video_data_numbers.csv"
uk_path = "./youtube_video_data/GB_video_data_numbers.csv"
us_data = np.loadtxt(us_path, delimiter=",", dtype=np.int64)
uk_data = np.loadtxt(uk_path, delimiter=",", dtype=np.int64)
# 提取us_data和uk_data的第一列作为views
max_view = 20000000
us_views, uk_views = us_data[:1600, 0], uk_data[:, 0]
us_views, uk_views = [v[v <= max_view] for v in (us_views, uk_views)]
# 绘制散点图，将US和UK数据用不同颜色绘制在同一图中
plt.scatter(range(len(us_views)), us_views, color="orange", label="US")
plt.scatter(range(len(uk_views)), uk_views, color="blue", label="UK")
plt.legend()

plt.title("US vs UK Video Views")
plt.show()
```

### 8、完成nan和inf的使用，掌握数组中nan值去除，替换为对应列的平均值

```python
a = np.array([[1, 2, np.nan, 4, 5], [6, np.nan, 7, 8, 9]])
col_mean = np.nanmean(a, axis=0)
col_mean = np.repeat(col_mean.reshape(1, -1), a.shape[0], axis=0)
print(col_mean)
a = np.where(np.isnan(a), col_mean, a)
print(a)
```
