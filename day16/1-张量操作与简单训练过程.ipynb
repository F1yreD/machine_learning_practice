# 张量的数值运算

```python
a = torch.ones(2, 3, dtype=torch.float)
print(a)

b = torch.full((2, 3), 2, dtype=torch.float)
print(b)

# 1. 直接与数字进行操作，会对张量进行广播
print("a + 10:", a + 10)
print("a - 10:", a - 10)
print("a * 10:", a * 10)
print("a / 10:", a / 10)

# 2. 张量之间的加减乘除(注意：需要形状一致)
print("a + b:", a + b)
print("a - b:", a - b)
print("a * b:", a * b)
print("a / b:", a / b)

print('--------------------------------')
# 3. 使用函数(不改变原张量)
print("a.add(b):", a.add(b))
print("a.sub(b):", a.sub(b))
print("a.mul(b):", a.mul(b))
print("a.div(b):", a.div(b))
print(a)
print('--------------------------------')
# 4. 使用函数(改变原张量)
print("a.add_(b):", a.add_(b))
# print("a.sub_(b):", a.sub_(b))
# print("a.mul_(b):", a.mul_(b))
# print("a.div_(b):", a.div_(b))
print(a)
```

```python
#矩阵乘法
a = torch.randint(0, 10, (2,3))
b = torch.randint(0, 10, (3,2))
print("a.mm(b):", a.mm(b))              # 只支持2维张量
print("a.matmul(b):", a.matmul(b))      # 支持任意维度张量
print("a @ b:", a @ b)                  # 支持任意维度张量
```

```python
import torch

x = torch.randn(2, 3, 4)
print("{获取张量形状}:", x.shape)
print("{获取张量大小}:", x.size())         # size() 方法和 shape 一样，返回张量的形状

# 1. 调整形状
x = x.reshape(2, 12)            # 调整为2行12列
print("{调整为2行12列}:",x.shape)
x = x.reshape(-1, 3)                # 不限制行数，列数调整为3列（会自动计算行数）
print("{自动计算行数，3列}:",x.shape)


# 2. 交换维度
x = torch.randn(2, 3, 4)
x = x.transpose(1, 2)
print("{交换第1和第2维度}:", x.shape)

# 3. 重新排列多个维度
# 第3维度变为第1维，第2维变为第2维，第1维变为第3维
x = x.permute(2,1,0)
print("{重新排列维度(2,1,0)}:", x.shape)

# 4. 在指定维度上增加一个维度， 会添加形状为1的维度
x = x.unsqueeze(dim = 2)
print("{在第2维增加维度}:", x.shape)

# 5. 在指定维度上删除一个维度， 会删除形状为1的维度（如果该维度不存在形状为1的维度，则返回原张量）
x = x.squeeze(dim = 2)
print("{删除第1维的维度}:", x.shape)

# 6. 调整形状，需要连续内存
# 经过升维和降维操作，张量的数据存储位置（内存）可能会改变，所以内存可能不连续
# 使用view方法，如果内存不连续，那么会报错
# x = x.view(3,-1)
# print(f'view后 {x.shape}')

# 6. 判断是否是连续内存
print("{判断内存是否连续}:", x.is_contiguous())

# 7. 转换为连续内存
x = x.contiguous()
print("{转换后内存是否连续}:", x.is_contiguous())

# 8. 调整形状，需要连续内存
x = x.view(3,-1)
print("{使用view调整形状}:", x.shape)
```

# view的数据共享问题

```python
# 9. 验证view操作是否共享内存
x = torch.randn(2, 3, 4)
print("{原始张量形状}:", x.shape)
print("{原始张量}:\n", x)

# 使用view调整形状
x_view = x.view(6, 4)
print("{view后张量形状}:", x_view.shape)

# 验证是否共享内存 - 修改原张量
x[0, 0, 0] = 999
print("{修改原张量后，原张量第一个元素}:", x[0, 0, 0])
print("{修改原张量后，view张量第一个元素}:", x_view[0, 0])

# 验证是否共享内存 - 修改view张量
x_view[1, 1] = 888
print("{修改view张量后，view张量[1,1]}:", x_view[1, 1])
print("{修改view张量后，原张量对应位置}:", x[0, 1, 1])

# 检查内存地址是否相同
print("{原张量id}:", id(x))
print("{view张量id}:", id(x_view))

print("{原张量数据指针}:", x.data_ptr())
print("{view张量数据指针}:", x_view.data_ptr())
print("{是否共享存储}:", x.storage().data_ptr() == x_view.storage().data_ptr())
```

# torch中到导数（梯度）计算

```python
import torch

# 1. 准备数据: 参数与目标值
x = torch.tensor(5) #特征
y = torch.tensor(0.)

# 2. 初始化参数: 权重与偏置
w = torch.tensor(1., requires_grad=True, dtype=torch.float32)
b = torch.tensor(3., requires_grad=True, dtype=torch.float32)

# 3. 构建神经网络并输出值
z = w * x + b

# 4. 计算损失
loss =  torch.nn.MSELoss()
loss = loss(z, y)

# 5. 计算梯度
loss.backward()

# 6. 访问梯度
print(w.grad)
print(b.grad)
```

```python
# 线性回归案例

import torch
import torch.nn as nn
import torch.optim as optim
from  torch.utils.data import TensorDataset
from  torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# 1. 生成数据
# 100个样本，1个特征，噪声为10，bias为1.5, coef=True: 返回基础线性回归的系数
x,y,coef = make_regression(n_samples=100, n_features=1, noise=10, bias=1.5, coef=True, random_state=22)

# 2. 数据可视化
# plt.scatter(x,y)

# 3. 使用Pytorch模拟线性回归
# 3.1 构造数据集对象 Dataset
dataset = TensorDataset(torch.from_numpy(x), torch.from_numpy(y))

# 3.2 创建数据加载器 DataLoader
# shuffle=:样本数据是否进行乱序
# batch_size: 每次迭代的样本数量
dataLoader = DataLoader(dataset, batch_size=10, shuffle=True)

# 3.3 构造模型
model = nn.Linear(in_features=1, out_features=1)

# 3.4 构造损失函数, 优化函数
loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# 3.5 训练模型
epochs = 100        # 迭代次数
losses = []         # 损失值列表
total_loss = 0.     # 损失值
train_sample = 0.   # 训练样本数

for epoch in range(epochs):
    for train_x, train_y in dataLoader:

        # 训练模型，预测结果（注意类型转换）
        y_pred = model(train_x.type(torch.float32))

        # 计算损失（注意类型转换）
        loss = loss_fn(y_pred, train_y.reshape(-1,1).type(torch.float32))

        # 反向传播，更新参数
        optimizer.zero_grad()       # 清空梯度
        loss.backward()             # 反向传播
        optimizer.step()            # 更新参数

        # print("epoch:{}, loss:{}".format(epoch, loss.item()))
        total_loss += loss.item()
        train_sample += len(train_y)

    # 单个epoch的平均训练损失
    losses.append(total_loss / train_sample)
# 4. 绘制损失变化曲线
plt.plot(range(epochs), losses)
plt.title("loss curve")
plt.show()

# 5. 绘制拟合曲线
plt.scatter(x,y)            # 数据散点图

x = torch.linspace(x.min(), x.max(), 100)
y_train = torch.tensor([v*model.weight + model.bias for v in x])        # 使用训练的权重和偏置
y_true = torch.tensor([v*coef + 1.5 for v in x])                        # 使用真实的权重和偏置

# 绘制曲线
plt.plot(x, y_train, label = 'train')
plt.plot(x, y_true, label = 'true')
plt.grid()                              # 绘制网格
plt.legend()                            # 绘制图例
plt.show()
```
