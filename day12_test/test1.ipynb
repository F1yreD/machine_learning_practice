### 1、机器学习的DictVectorizer练习，文本分词，CountVectorizer, TfidfVectorizer练习

```python
from sklearn.feature_extraction import DictVectorizer
import pandas as pd

data = [{'a':1,'b':2},
{'a':3,'b':4}
]

dict_vec = DictVectorizer(sparse=False)
X = dict_vec.fit_transform(data)
pd.DataFrame(X, columns=dict_vec.get_feature_names_out())
```

### 2、归一化，标准化，缺失值插补练习

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.impute import SimpleImputer
data = [[1,0],[2,20],[3,30]]
X = MinMaxScaler(feature_range=(0,1)).fit_transform(data)
print(X)

Y = StandardScaler(with_mean=False).fit_transform(data)
print(Y)

Z = SimpleImputer(missing_values=0,strategy='mean').fit_transform(data)
print(Z)
```

### 3、练习特征预处理中的VarianceThreshold和PCA

```python
from sklearn.feature_selection import VarianceThreshold
from sklearn.decomposition import PCA

data = [[1,2,3,4],[8,7,6,5]]
sel = VarianceThreshold(threshold=0.5)
X = sel.fit_transform(data)
print(X)

print(sel.variances_,sel.get_support())

pca = PCA(n_components=0.95)
Y = pca.fit_transform(data)
print(Y,Y.shape)
print(pca.explained_variance_ratio_)
```

### 4、完成对sklearn自带数据集的data，target，feature_names，target_names的理解

```python
from sklearn.datasets import load_iris
import pandas as pd

# 加载鸢尾花数据集
iris = load_iris()

# 创建DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

X=iris.data  #特征
y=iris.target #标签
# 输出数据的基本信息
print("数据的前5行：")
print(df.head())
print("\n数据的维度：", df.shape)
print("\n特征名称：", iris.feature_names)
print("\n标签种类：", iris.target_names)
#打印X与y的shape
print(X.shape)
print(y.shape)
print(y)
#打印X的数据类型
print(X.dtype)
#打印y的数据类型
print(y.dtype)
```

### 5、使用KNN完成鸢尾花的训练及评估，与上课一致。

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# 划分训练集和测试集，测试集占 30%,random_state随机种子，作用是每次运行代码时，划分结果都相同
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 创建KNN分类器实例，设定邻居数为3
knn = KNeighborsClassifier(n_neighbors=3)

# 用训练数据拟合模型
knn.fit(X_train, y_train)

# 在测试集上进行预测
y_pred = knn.predict(X_test)

# 输出预测结果
print("KNN预测的前10个结果：", y_pred[:10])

from sklearn.metrics import accuracy_score

# 计算测试集上的准确率
accuracy = accuracy_score(y_test, y_pred)
print("测试集上的准确率为：", accuracy)
```
