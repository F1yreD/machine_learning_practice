# 用knn算法对鸢尾花进行分类

```python
from sklearn.datasets import load_iris
import pandas as pd

# 加载鸢尾花数据集
iris = load_iris()

# 创建DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

X=iris.data  #特征
y=iris.target #标签
# 输出数据的基本信息
print("数据的前5行：")
print(df.head())
print("\n数据的维度：", df.shape)
print("\n特征名称：", iris.feature_names)
print("\n标签种类：", iris.target_names)
#打印X与y的shape
print(X.shape)
print(y.shape)
print(y)
#打印X的数据类型
print(X.dtype)
#打印y的数据类型
print(y.dtype)
```

```python
df.iloc[-5:,:]
```

```python
from sklearn.model_selection import train_test_split

# 划分训练集和测试集，测试集占 30%,random_state随机种子，作用是每次运行代码时，划分结果都相同
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 打印训练集与测试集的形状
print("训练集 X_train 形状：", X_train.shape)
print("测试集 X_test 形状：", X_test.shape)
print("训练集 y_train 形状：", y_train.shape)
print("测试集 y_test 形状：", y_test.shape)
X_train
```

# 训练及预测

```python

from sklearn.neighbors import KNeighborsClassifier

# 创建KNN分类器实例，设定邻居数为3
knn = KNeighborsClassifier(n_neighbors=3)

# 用训练数据拟合模型
knn.fit(X_train, y_train)

# 在测试集上进行预测
y_pred = knn.predict(X_test)

# 输出预测结果
print("KNN预测的前10个结果：", y_pred[:10])
```

```python
from sklearn.metrics import accuracy_score

# 计算测试集上的准确率
accuracy = accuracy_score(y_test, y_pred)
print("测试集上的准确率为：", accuracy)
```
