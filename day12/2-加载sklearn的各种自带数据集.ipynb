```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import pandas as pd

# 加载手写数字数据集
digits = load_digits()

# 展示数据集的基本信息
print("数据集的键:", digits.keys())
print("\n数据集描述（前100字符）:", digits['DESCR'][:100], "...\n")
print("数据集形状 (images):", digits.images.shape)
print("数据集形状 (data):", digits.data.shape)
print("特征数:", digits.data.shape[1])
print("标签类别:", digits.target_names)
print("标签 shape:", digits.target.shape)
print("首10个标签:", digits.target[:10])
print("首5条数据:\n", digits.data[:5])

plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号

# 显示部分手写图片以及对应标签
plt.figure(figsize=(10, 2))
for i in range(10):
    plt.subplot(1, 10, i + 1)
    plt.imshow(digits.images[i], cmap='gray') # 显示灰度图像,imshow()函数用于显示图像
    plt.title(f"标签:{digits.target[i]}")
    plt.axis('off')
plt.suptitle("前10个手写数字图片及其标签")
plt.show()

# 用DataFrame展示前几条样本及特征
df = pd.DataFrame(digits.data, columns=[f'pixel_{i}' for i in range(digits.data.shape[1])])
df['target'] = digits.target
print("\n前5条样本及标签（DataFrame 形式）:")
print(df.head())

```

```python
from sklearn.datasets import load_breast_cancer

# 加载乳腺癌数据集
cancer = load_breast_cancer()

# 展示乳腺癌数据集的基本信息
print("数据集的键:", cancer.keys())
print("\n数据集描述（前100字符）:", cancer['DESCR'][:100], "...\n")
print("数据集形状 (data):", cancer.data.shape)
print("特征数:", cancer.data.shape[1])
print("特征名称:", cancer.feature_names)
print("标签类别:", cancer.target_names)
print("标签 shape:", cancer.target.shape)
print("标签分布（每个类别的数量）:", pd.Series(cancer.target).value_counts())
print("首10个标签:", cancer.target[:10])
print("首5条数据（特征值）:\n", cancer.data[:5])

# 用DataFrame展示前几条样本及特征
df_cancer = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df_cancer['target'] = cancer.target
print("\n前5条样本及标签（DataFrame 形式）:")
print(df_cancer.head())
```

```python
from sklearn.datasets import fetch_california_housing

# 加载加利福尼亚房价数据集
california = fetch_california_housing(data_home='./data/')

# 展示数据集的键
print("数据集的键:", california.keys())

# 展示数据集的描述信息（前500字符）
print("\n数据集描述（前500字符）:\n", california['DESCR'][:500], "...\n")

# 特征数据形状
print("特征数据形状 (data):", california.data.shape)
print("特征数量:", california.data.shape[1])
print("特征名称:", california.feature_names)

# 标签数据形状
print("标签 shape:", california.target.shape)
print("首10个标签(y):", california.target[:10])

# 用 pandas 展示前几条样本及标签
df_california = pd.DataFrame(california.data, columns=california.feature_names)
df_california['target'] = california.target
print("\n前5条样本及标签（DataFrame 形式）:")
print(df_california.head())

# 标签直方图（简单分析标签分布）
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.hist(california.target, bins=40, color='skyblue', edgecolor='black')
plt.title('房价标签分布 (单位: 以万美元计)')
plt.xlabel('房价 (单位: 万美元)')
plt.ylabel('样本数量')
plt.show()

# 各个特征的基本统计信息
print("\n特征的描述性统计信息（前5行）:")
print(df_california.describe().T.head())

# 保存到当前目录
df_california.to_csv("california_housing.csv", index=False)
print("\n数据集已保存到当前目录下的 'california_housing.csv' 文件。")


print(california['DESCR'])
```
