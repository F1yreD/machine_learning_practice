```python
from sklearn.datasets import fetch_20newsgroups

# 加载20类新闻数据集
newsgroups = fetch_20newsgroups(data_home='../day14/data',subset='all')

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(
    newsgroups.data, newsgroups.target, test_size=0.25, random_state=42)

# 对训练集进行TF-IDF向量化
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# 输出处理后的特征数
print("处理后的特征数：", X_train_tfidf.shape[1])
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

# 朴素贝叶斯训练
nb = MultinomialNB(alpha=1)
nb.fit(X_train_tfidf, y_train)

# 预测
y_pred = nb.predict(X_test_tfidf)

# 评估
print("准确率：", accuracy_score(y_test, y_pred))
```

```python
def precision_score(y_true, y_pred):
    tp = sum(1 for t, p in zip(y_true, y_pred) if t == 1 and p == 1)  # 真正例
    fp = sum(1 for t, p in zip(y_true, y_pred) if t == 0 and p == 1)  # 假正例
    
    if tp + fp == 0:
        return 0.0
    
    precision = tp / (tp + fp)
    return precision

alt_atheism_index = newsgroups.target_names.index('alt.atheism')
print(y_pred)

# 预测结果中被分类为 alt.atheism 的样本索引
predicted_as_alt_atheism = (y_pred == alt_atheism_index)

# 实际标签中是 alt.atheism 的样本索引
actual_alt_atheism = (y_test == alt_atheism_index)
precision_atheism = precision_score(actual_alt_atheism, predicted_as_alt_atheism)
print("alt.atheism 类别的精确率:", precision_atheism)
```
