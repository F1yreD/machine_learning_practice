```python
# 例子：理解面向对象中的 __call__

class MyCallable:
    def __init__(self, prefix):
        self.prefix = prefix

    def __call__(self, name):
        print(f"{self.prefix}, {name}!")

# 创建对象
greeter = MyCallable("你好")
# 对象像函数一样被调用
greeter("世界")  # 输出: 你好, 世界!

# 也可以创建多个带有不同前缀的实例
eng_greeter = MyCallable("Hello")
eng_greeter("world")  # 输出: Hello, world!
```
