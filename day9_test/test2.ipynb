### 折线图

```python
from matplotlib import pyplot

x = range(1, 11)
y = range(11, 1, -1)
pyplot.plot(x, y, color='red', alpha=0.5, linestyle='-.', linewidth=2)
x_ticks = [f"{i}.00" for i in x]
pyplot.xticks(x,x_ticks,rotation = 90)
pyplot.show()
```

### 散点图

```python
import random
x = [random.randint(1, 10) for _ in range(10)]
y = [random.randint(1, 10) for _ in range(10)]
pyplot.scatter(x, y,c='red', alpha=0.5, s=50,marker='s')
pyplot.show()
```

### 条形图

```python
x = ['A', 'B', 'C', 'D']
y = [random.randint(1, 10) for _ in range(4)]
pyplot.bar(x, y, color='red', alpha=0.5)
pyplot.show()
```

### 饼图

```python
x = ['A', 'B', 'C', 'D']
y = [random.randint(1, 10) for _ in range(4)]
pyplot.pie(y, labels=x, autopct='%.2f%%')
pyplot.show()
```
